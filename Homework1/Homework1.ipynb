***

The founding of the University of Texas was a long and often-delayed process, born from a grand vision that took nearly 50 years to become a reality. Its story is one of constitutional ambition, political battles, humble beginnings, and a dramatic, fortune-changing discovery.

### 1. The Constitutional Dream (1839)
The story begins not with a building, but with an idea. In 1839, the Congress of the Republic of Texas, urged by President Mirabeau B. Lamar, set aside 50 leagues of land (about 221,400 acres) for the establishment of "a university of the first class." This act embedded the creation of a premier public university into the very foundation of Texas. However, due to frontier conflicts, the financial struggles of the Republic, and the impending U.S. Civil War and Reconstruction, this dream remained deferred for decades.

### 2. The Mandate Renewed (1876)
The vision was officially revived in the **Texas Constitution of 1876**. This new constitution mandated that the state legislature establish the University of Texas "as soon as practicable." It also created the **Permanent University Fund (PUF)**, a public endowment, and dedicated one million acres of land in West Texas to it. This land, considered poor quality for farming or ranching, was thought to be of little value at the time.

### 3. The Fight for Location (1881)
With the constitutional mandate in place, the next step was deciding *where* to build the university. In 1881, a statewide election was held to determine the location. Cities across Texas, including Waco, Tyler, and Lampasas, vied for the honor.

Ultimately, **Austin** was chosen for the main university, while **Galveston** was selected to be the home of the medical department.

### 4. A Humble Opening (1883)
After nearly half a century of planning and political maneuvering, the University of Texas officially opened its doors on **September 15, 1883**. The beginning was modest. Classes were held in a single, temporary building on a 40-acre campus known as "College Hill." The university began with just **eight professors, one proctor, and 221 students**. The first permanent structure, the Victorian-Gothic Main Building, was completed a few years later.

### 5. The Black Gold Transformation (1923)
For its first 40 years, the university was constantly short on funds, relying on the seemingly worthless land of the Permanent University Fund. The story took a dramatic turn on **May 28, 1923**. An oil well named **Santa Rita No. 1** struck oil on university-owned land in West Texas.

This discovery transformed the university’s fate overnight. The "worthless" land began generating immense wealth, turning the PUF into one of the largest educational endowments in the world. This oil revenue allowed the University of Texas to fund the construction of its campus and fulfill its founding mission to become a "university of the first class."
